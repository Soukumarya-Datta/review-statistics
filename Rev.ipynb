```python
def previous_3_months_time():
    time = {}
    rev = {}
    df = pd.DataFrame()
    for i in data_23:
        if(i['closeDate'] > "2023-09-31" and i['closeDate'] <= "2023-12-31"):
            for j in i['reviewers']['reviewer']:
                if "timeSpent" in j:
                    time[j['userName']] = 0
                    rev[j['userName']] = 0
        
    for i in data_23:
        if(i['closeDate'] > "2023-09-31" and i['closeDate'] <= "2023-12-31"):
            for j in i['reviewers']['reviewer']:
                if "timeSpent" in j:
                    time[j['userName']] = time[j['userName']] + round(j['timeSpent']/3600000,2)
                    rev[j['userName']] = rev[j['userName']] + 1
        
    df_t = pd.DataFrame(list(time.items()), columns=['user', 'timeSpent'])
    df_r = pd.DataFrame(list(rev.items()), columns=['user', 'No. of reviews'])
    df_t = pd.merge(df_t, df_r, on='user')
    
    for j in data_23:
        if(j['closeDate'] > "2023-09-31" and j['closeDate'] <= "2023-12-31"):
            for i in j['stats']:
                df = df.append(i, ignore_index=True)
            
    df = df.drop(columns=['drafts', 'unread', 'leaveUnread', 'read'])
    df = df.groupby('user').sum().reset_index()
    df = pd.merge(df, df_t, on='user')
    
    global def_qtr
    def_qtr = df
    def_qtr = def_qtr.rename(columns={"user" : "Asst. ID", "published" : "Comments", "defects" : "Defects", "timeSpent" : "Time spent (hrs)"})
    def_qtr = def_qtr.sort_values(by='Defects', ascending=False)
    return df

def previous_6_months_time():
    previous_3_months_time()
    time = {}
    rev = {}
    df = pd.DataFrame()
    for i in data_23:
        if(i['closeDate'] > "2023-06-30" and i['closeDate'] <= "2023-12-31"):
            for j in i['reviewers']['reviewer']:
                if "timeSpent" in j:
                    time[j['userName']] = 0
                    rev[j['userName']] = 0
        
    for i in data_23:
        if(i['closeDate'] > "2023-06-30" and i['closeDate'] <= "2023-12-31"):
            for j in i['reviewers']['reviewer']:
                if "timeSpent" in j:
                    time[j['userName']] = time[j['userName']] + round(j['timeSpent']/3600000,2)
                    rev[j['userName']] = rev[j['userName']] + 1
        
    df_t = pd.DataFrame(list(time.items()), columns=['user', 'timeSpent'])
    df_r = pd.DataFrame(list(rev.items()), columns=['user', 'No. of reviews'])
    df_t = pd.merge(df_t, df_r, on='user')
    
    for j in data_23:
        if(j['closeDate'] > "2023-06-30" and j['closeDate'] <= "2023-12-31"):
            for i in j['stats']:
                df = df.append(i, ignore_index=True)
            
    df = df.drop(columns=['drafts', 'unread', 'leaveUnread', 'read'])
    df = df.groupby('user').sum().reset_index()
    df = pd.merge(df, df_t, on='user')
    
    global def_half
    def_half = df
    def_half = def_half.rename(columns={"user" : "Asst. ID", "published" : "Comments", "defects" : "Defects", "timeSpent" : "Time spent (hrs)"})
    def_half = def_half.sort_values(by='Defects', ascending=False)
    return df

def previous_year_time():
    previous_6_months_time()
    time = {}
    rev = {}
    df = pd.DataFrame()
    for i in data_23:
        for j in i['reviewers']['reviewer']:
            if "timeSpent" in j:
                time[j['userName']] = 0
                rev[j['userName']] = 0
        
    for i in data_23:
        for j in i['reviewers']['reviewer']:
            if "timeSpent" in j:
                time[j['userName']] = time[j['userName']] + round(j['timeSpent']/3600000,2)
                rev[j['userName']] = rev[j['userName']] + 1
        
    df_t = pd.DataFrame(list(time.items()), columns=['user', 'timeSpent'])
    df_r = pd.DataFrame(list(rev.items()), columns=['user', 'No. of reviews'])
    df_t = pd.merge(df_t, df_r, on='user')
    
    # For calculating Number of comments and defects
    for j in data_23:
        for i in j['stats']:
            df = df.append(i, ignore_index=True)
            
    df = df.drop(columns=['drafts', 'unread', 'leaveUnread', 'read'])
    df = df.groupby('user').sum().reset_index()
    df = pd.merge(df, df_t, on='user')
    
    global def_23
    def_23 = df
    def_23 = def_23.rename(columns={"user" : "Asst. ID", "published" : "Comments", "defects" : "Defects", "timeSpent" : "Time spent (hrs)"})
    def_23 = def_23.sort_values(by='Defects', ascending=False)
    return df

def previous_two_year_time():
    time = {}
    rev = {}
    df = pd.DataFrame()
    df_yr = previous_year_time()
    for i in data_22:
        for j in i['reviewers']['reviewer']:
            if "timeSpent" in j:
                time[j['userName']] = 0
                rev[j['userName']] = 0
        
    for i in data_22:
        for j in i['reviewers']['reviewer']:
            if "timeSpent" in j:
                time[j['userName']] = time[j['userName']] + round(j['timeSpent']/3600000,2)
                rev[j['userName']] = rev[j['userName']] + 1
    
    
    df_t = pd.DataFrame(list(time.items()), columns=['user', 'timeSpent'])
    df_r = pd.DataFrame(list(rev.items()), columns=['user', 'No. of reviews'])
    df_t = pd.merge(df_t, df_r, on='user')
    
    merged_df = pd.merge(df_t, df_yr, on='user', how='outer')
    merged_df.fillna(0.0, inplace = True)
    merged_df["timeSpent"] = merged_df["timeSpent_x"] + merged_df["timeSpent_y"]
    merged_df.drop(columns=["timeSpent_x", "timeSpent_y"], inplace = True)
    
    # For calculating Number of comments and defects
    for j in data_22:
        for i in j['stats']:
            df = df.append(i, ignore_index=True)
            
    df = df.drop(columns=['drafts', 'unread', 'leaveUnread', 'read'])
    df = df.groupby('user').sum().reset_index()
    df = pd.merge(df, merged_df, on='user')
    df.fillna(0.0, inplace = True)
    df["published"] = df["published_x"] + df["published_y"]
    df["defects"] = df["defects_x"] + df["defects_y"]
    df["No. of reviews"] = df["No. of reviews_x"] + df["No. of reviews_y"]
    df.drop(columns=["published_x", "published_y", "defects_x", "defects_y", "No. of reviews_x", "No. of reviews_y"], inplace = True)
    
    global def_22
    def_22 = df
    def_22 = def_22.rename(columns={"user" : "Asst. ID", "published" : "Comments", "defects" : "Defects", "timeSpent" : "Time spent (hrs)"})
    def_22 = def_22.sort_values(by='Defects', ascending=False)
    return df

def previous_three_year_time():
    time = {}
    rev = {}
    df = pd.DataFrame()
    df_yr = previous_two_year_time()
    for i in data_21:
        for j in i['reviewers']['reviewer']:
            if "timeSpent" in j:
                time[j['userName']] = 0
                rev[j['userName']] = 0
        
    for i in data_21:
        for j in i['reviewers']['reviewer']:
            if "timeSpent" in j:
                time[j['userName']] = time[j['userName']] + round(j['timeSpent']/3600000,2)
                rev[j['userName']] = rev[j['userName']] + 1
        
    df_t = pd.DataFrame(list(time.items()), columns=['user', 'timeSpent'])
    df_r = pd.DataFrame(list(rev.items()), columns=['user', 'No. of reviews'])
    df_t = pd.merge(df_t, df_r, on='user')
    merged_df = pd.merge(df_t, df_yr, on='user', how='outer')
    merged_df.fillna(0.0, inplace = True)
    merged_df["timeSpent"] = merged_df["timeSpent_x"] + merged_df["timeSpent_y"]
    merged_df.drop(columns=["timeSpent_x", "timeSpent_y"], inplace = True)
    
    # For calculating Number of comments and defects
    for j in data_21:
        for i in j['stats']:
            df = df.append(i, ignore_index=True)
            
    df = df.drop(columns=['drafts', 'unread', 'leaveUnread', 'read'])
    df = df.groupby('user').sum().reset_index()
    df = pd.merge(df, merged_df, on='user')
    df.fillna(0.0, inplace = True)
    df["published"] = df["published_x"] + df["published_y"]
    df["defects"] = df["defects_x"] + df["defects_y"]
    df["No. of reviews"] = df["No. of reviews_x"] + df["No. of reviews_y"]
    df.drop(columns=["published_x", "published_y", "defects_x", "defects_y", "No. of reviews_x", "No. of reviews_y"], inplace = True)
    
    global def_21
    def_21 = df
    def_21 = def_21.rename(columns={"user" : "Asst. ID", "published" : "Comments", "defects" : "Defects", "timeSpent" : "Time spent (hrs)"})
    def_21 = def_21.sort_values(by='Defects', ascending=False)
    return df

def previous_four_year_time():
    time = {}
    rev = {}
    df = pd.DataFrame()
    df_yr = previous_three_year_time()
    for i in data_20:
        for j in i['reviewers']['reviewer']:
            if "timeSpent" in j:
                time[j['userName']] = 0
                rev[j['userName']] = 0
        
    for i in data_20:
        for j in i['reviewers']['reviewer']:
            if "timeSpent" in j:
                time[j['userName']] = time[j['userName']] + round(j['timeSpent']/3600000,2)
                rev[j['userName']] = rev[j['userName']] + 1
        
    df_t = pd.DataFrame(list(time.items()), columns=['user', 'timeSpent'])
    df_r = pd.DataFrame(list(rev.items()), columns=['user', 'No. of reviews'])
    df_t = pd.merge(df_t, df_r, on='user')
    merged_df = pd.merge(df_t, df_yr, on='user')
    merged_df.fillna(0.0, inplace = True)
    merged_df["timeSpent"] = merged_df["timeSpent_x"] + merged_df["timeSpent_y"]
    merged_df.drop(columns=["timeSpent_x", "timeSpent_y"], inplace = True)
    
    # For calculating Number of comments and defects
    for j in data_20:
        for i in j['stats']:
            df = df.append(i, ignore_index=True)
            
    df = df.drop(columns=['drafts', 'unread', 'leaveUnread', 'read'])
    df = df.groupby('user').sum().reset_index()
    df = pd.merge(df, merged_df, on='user')
    df.fillna(0.0, inplace = True)
    df["published"] = df["published_x"] + df["published_y"]
    df["defects"] = df["defects_x"] + df["defects_y"]
    df["No. of reviews"] = df["No. of reviews_x"] + df["No. of reviews_y"]
    df.drop(columns=["published_x", "published_y", "defects_x", "defects_y", "No. of reviews_x", "No. of reviews_y"], inplace = True)
    
    global def_20
    def_20 = df
    def_20 = def_20.rename(columns={"user" : "Asst. ID", "published" : "Comments", "defects" : "Defects", "timeSpent" : "Time spent (hrs)"})
    def_20 = def_20.sort_values(by='Defects', ascending=False)
    return df

df = previous_four_year_time()
df.head()
```

```python
%%time
def tables(yr):
    if(yr=='Last Quater'):
        def_qtr["Time spent (hrs)"] = def_qtr["Time spent (hrs)"].round(2)
        return def_qtr
    if(yr=='Past 6 months'):
        def_half["Time spent (hrs)"] = def_half["Time spent (hrs)"].round(2)
        return def_half            
    if(yr=='Since 2023'):
        def_23["Time spent (hrs)"] = def_23["Time spent (hrs)"].round(2)
        return def_23            
    if(yr=='Since 2022'):
        def_22["Time spent (hrs)"] = def_22["Time spent (hrs)"].round(2)
        return def_22            
    if(yr=='Since 2021'):
        def_21["Time spent (hrs)"] = def_21["Time spent (hrs)"].round(2)
        return def_21            
    if(yr=='Since 2020'):
        def_20["Time spent (hrs)"] = def_20["Time spent (hrs)"].round(2)
        return def_20
    
    
def associate(yr, aid=None):
    df = tables(yr)
    if(aid==None):
        return df
    else:
        return df[df["Asst. ID"] == aid]
    

def Plots(yr, stat):
    if(yr=='Last Quater'):
        if(stat=='Defects'):
            def_qtr.sort_values(by='Defects', ascending=False, inplace = True)
            fig = px.bar(def_qtr, x='Asst. ID', y=def_qtr["Defects"])
            fig.update_layout(title="Defects", yaxis_title="No. of Defects", xaxis_title="Associate")
        if(stat=='Comments'):
            def_qtr.sort_values(by='Comments', ascending=False, inplace = True)
            fig = px.bar(def_qtr, x='Asst. ID', y=def_qtr["Comments"])
            fig.update_layout(title="Comments", yaxis_title="No. of Defects", xaxis_title="Associate")
        if(stat=='Time Spent'):
            def_qtr.sort_values(by='Time spent (hrs)', ascending=False, inplace = True)
            fig = px.bar(def_qtr, x='Asst. ID', y=def_qtr["Time spent (hrs)"])
            fig.update_layout(title="Time Spent", yaxis_title="Hours", xaxis_title="Associate")
        if(stat=='No. of reviews'):
            def_qtr.sort_values(by='No. of reviews', ascending=False, inplace = True)
            fig = px.bar(def_qtr, x='Asst. ID', y=def_qtr["No. of reviews"])
            fig.update_layout(title="No. of reviews", yaxis_title="Reviews", xaxis_title="Associate")
            
    if(yr=='Past 6 months'):
        if(stat=='Defects'):
            def_half.sort_values(by='Defects', ascending=False, inplace = True)
            fig = px.bar(def_half, x='Asst. ID', y=def_half["Defects"])
            fig.update_layout(title="Defects", yaxis_title="No. of Defects", xaxis_title="Associate")
        if(stat=='Comments'):
            def_half.sort_values(by='Comments', ascending=False, inplace = True)
            fig = px.bar(def_half, x='Asst. ID', y=def_half["Comments"])
            fig.update_layout(title="Comments", yaxis_title="No. of Defects", xaxis_title="Associate")
        if(stat=='Time Spent'):
            def_half.sort_values(by='Time spent (hrs)', ascending=False, inplace = True)
            fig = px.bar(def_half, x='Asst. ID', y=def_half["Time spent (hrs)"])
            fig.update_layout(title="Time Spent", yaxis_title="Hours", xaxis_title="Associate")
        if(stat=='No. of reviews'):
            def_half.sort_values(by='No. of reviews', ascending=False, inplace = True)
            fig = px.bar(def_half, x='Asst. ID', y=def_half["No. of reviews"])
            fig.update_layout(title="No. of reviews", yaxis_title="Reviews", xaxis_title="Associate")
            
    if(yr=='Since 2023'):
        if(stat=='Defects'):
            def_23.sort_values(by='Defects', ascending=False, inplace = True)
            fig = px.bar(def_23, x='Asst. ID', y=def_23["Defects"])
            fig.update_layout(title="Defects", yaxis_title="No. of Defects", xaxis_title="Associate")
        if(stat=='Comments'):
            def_23.sort_values(by='Comments', ascending=False, inplace = True)
            fig = px.bar(def_23, x='Asst. ID', y=def_23["Comments"])
            fig.update_layout(title="Comments", yaxis_title="No. of Defects", xaxis_title="Associate")
        if(stat=='Time Spent'):
            def_23.sort_values(by='Time spent (hrs)', ascending=False, inplace = True)
            fig = px.bar(def_23, x='Asst. ID', y=def_23["Time spent (hrs)"])
            fig.update_layout(title="Time Spent", yaxis_title="Hours", xaxis_title="Associate")
        if(stat=='No. of reviews'):
            def_23.sort_values(by='No. of reviews', ascending=False, inplace = True)
            fig = px.bar(def_23, x='Asst. ID', y=def_23["No. of reviews"])
            fig.update_layout(title="No. of reviews", yaxis_title="Reviews", xaxis_title="Associate")
            
    if(yr=='Since 2022'):
        if(stat=='Defects'):
            def_22.sort_values(by='Defects', ascending=False, inplace = True)
            fig = px.bar(def_22, x='Asst. ID', y=def_22["Defects"])
            fig.update_layout(title="Defects", yaxis_title="No. of Defects", xaxis_title="Associate")
        if(stat=='Comments'):
            def_22.sort_values(by='Comments', ascending=False, inplace = True)
            fig = px.bar(def_22, x='Asst. ID', y=def_22["Comments"])
            fig.update_layout(title="Comments", yaxis_title="No. of Defects", xaxis_title="Associate")
        if(stat=='Time Spent'):
            def_22.sort_values(by='Time spent (hrs)', ascending=False, inplace = True)
            fig = px.bar(def_22, x='Asst. ID', y=def_22["Time spent (hrs)"])
            fig.update_layout(title="Time Spent", yaxis_title="Hours", xaxis_title="Associate")
        if(stat=='No. of reviews'):
            def_22.sort_values(by='No. of reviews', ascending=False, inplace = True)
            fig = px.bar(def_22, x='Asst. ID', y=def_22["No. of reviews"])
            fig.update_layout(title="No. of reviews", yaxis_title="Reviews", xaxis_title="Associate")
            
    if(yr=='Since 2021'):
        if(stat=='Defects'):
            def_21.sort_values(by='Defects', ascending=False, inplace = True)
            fig = px.bar(def_21, x='Asst. ID', y=def_21["Defects"])
            fig.update_layout(title="Defects", yaxis_title="No. of Defects", xaxis_title="Associate")
        if(stat=='Comments'):
            def_21.sort_values(by='Comments', ascending=False, inplace = True)
            fig = px.bar(def_21, x='Asst. ID', y=def_21["Comments"])
            fig.update_layout(title="Comments", yaxis_title="No. of Defects", xaxis_title="Associate")
        if(stat=='Time Spent'):
            def_21.sort_values(by='Time spent (hrs)', ascending=False, inplace = True)
            fig = px.bar(def_21, x='Asst. ID', y=def_21["Time spent (hrs)"])
            fig.update_layout(title="Time Spent", yaxis_title="Hours", xaxis_title="Associate")
        if(stat=='No. of reviews'):
            def_21.sort_values(by='No. of reviews', ascending=False, inplace = True)
            fig = px.bar(def_21, x='Asst. ID', y=def_21["No. of reviews"])
            fig.update_layout(title="No. of reviews", yaxis_title="Reviews", xaxis_title="Associate")
            
    if(yr=='Since 2020'):
        if(stat=='Defects'):
            def_20.sort_values(by='Defects', ascending=False, inplace = True)
            fig = px.bar(def_20, x='Asst. ID', y=def_20["Defects"])
            fig.update_layout(title="Defects", yaxis_title="No. of Defects", xaxis_title="Associate")
        if(stat=='Comments'):
            def_20.sort_values(by='Comments', ascending=False, inplace = True)
            fig = px.bar(def_20, x='Asst. ID', y=def_20["Comments"])
            fig.update_layout(title="Comments", yaxis_title="No. of Defects", xaxis_title="Associate")
        if(stat=='Time Spent'):
            def_20.sort_values(by='Time spent (hrs)', ascending=False, inplace = True)
            fig = px.bar(def_20, x='Asst. ID', y=def_20["Time spent (hrs)"])
            fig.update_layout(title="Time Spent", yaxis_title="Hours", xaxis_title="Associate")
        if(stat=='No. of reviews'):
            def_20.sort_values(by='No. of reviews', ascending=False, inplace = True)
            fig = px.bar(def_20, x='Asst. ID', y=def_20["No. of reviews"])
            fig.update_layout(title="No. of reviews", yaxis_title="Reviews", xaxis_title="Associate")
    
    return fig


with gr.Blocks() as Tabular:
    inputs = [gr.Dropdown(["Since 2020", "Since 2021", "Since 2022", "Since 2023", "Past 6 months", "Last Quarter"], 
                         label="Year"),
             gr.Textbox(label="Associate ID", placeholder="(in UPPERCASE)")]
    out = gr.DataFrame()
    greet_btn = gr.Button("Search")
    inputs[0].change(tables, inputs[0], out)
    greet_btn.click(fn=associate, inputs=inputs, outputs=out, api_name="search")
    
with gr.Blocks() as Graphical:
    inputs = [gr.Dropdown(["Since 2020", "Since 2021", "Since 2022", "Since 2023", "Past 6 months", "Last Quarter"], 
                          label="Year"),
              gr.Radio(["Defects", "Comments", "Time Spent", "No. of reviews"], label="Review statictics")]
    out = gr.Plot()
    inputs[0].change(Plots, inputs, out)
    inputs[1].change(Plots, inputs, out)
    
demo = gr.TabbedInterface([Tabular, Graphical], ["Tabular", "Graphical"])

demo.close()
demo.launch()
```
